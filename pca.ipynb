## PCA - Principal Component Analysis

### Example Dataset

```python
# IMPORTS
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Dataset from Hands-On Machine Learning with Scikit-Learn, A. Geron
np.random.seed(4)
m = 60
w1, w2 = 0.1, 0.3
noise = 0.1

angles = np.random.rand(m) * 3 * np.pi / 2 - 0.5
X = np.empty((m, 3))
X[:, 0] = np.cos(angles) + np.sin(angles)/2 + noise * np.random.randn(m) / 2
X[:, 1] = np.sin(angles) * 0.7 + noise * np.random.randn(m) / 2
X[:, 2] = X[:, 0] * w1 + X[:, 1] * w2 + noise * np.random.randn(m)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(X[:,0],X[:,1] ,X[:,2],'r.')
```

## PCA with SVD 

```python
def pca_with_svd(X,n_vectors):

  X_centered=X-X.mean(axis=0)
  U,s,Vt= np.linalg.svd(X_centered)

  if n_vectors> (Vt.shape[1] +1):
    print('Invalid number of vectors, data has fewer dimensions')

  return X_centered,Vt[:,:n_vectors]
```

```python
X_centered,W=pca_with_svd(X,2)
```

```python
X_transformed=X_centered.dot(W)
```

```python
plt.plot(X_transformed[:,0],X_transformed[:,1],'.r' )
```
