Linear Regression - is a linear approach to modelling the relationship between a scalar response and one or more explanatory variables

image.png

```python
import numpy as np
import matplotlib.pyplot as plt

class LinearRegressor:

  def __init__(self,lr=1e-4,n_iters=1000):
    self.learning_rate=lr
    self.n_iters=n_iters
    self.weights=None
    self.bias=None

 
  def fit(self, X, y):
      n_samples, n_features = X.shape

      # init parameters
      self.weights = np.zeros(n_features)
      self.bias = 0

        # gradient descent
      for _ in range(self.n_iters):
          y_predicted = np.dot(X, self.weights) + self.bias
          # compute gradients
          dw = (1 / n_samples) * np.dot(X.T, (y_predicted - y))
          db = (1 / n_samples) * np.sum(y_predicted - y)

            # update parameters
          self.weights -= self.learning_rate * dw
          self.bias -= self.learning_rat * db

  def predict(self,X):
    
    return np.dot(X,self.weights) + self.bias
    
```
